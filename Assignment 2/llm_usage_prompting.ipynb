# LLM Usage: Zero-shot vs Few-shot + Chain-of-Thought (CoT)


## Part A — Zero-shot vs Few-shot Prompting



```python

ZERO_SHOT_PROMPT = """
You are a support-ticket classifier.
Classify the ticket into exactly one of: Billing, Technical, Account.
Return ONLY the label.

Ticket: "I was charged twice for my subscription this month. Please fix it."
"""

FEW_SHOT_PROMPT = """
You are a support-ticket classifier.
Classify the ticket into exactly one of: Billing, Technical, Account.
Return ONLY the label.

Ticket: "My password reset link keeps expiring."
Label: Account

Ticket: "The app crashes every time I upload a file."
Label: Technical

Ticket: "I was charged twice for my subscription this month. Please fix it."
Label:
"""

print("ZERO-SHOT PROMPT:\n")
print(ZERO_SHOT_PROMPT)
print("\nFEW-SHOT PROMPT:\n")
print(FEW_SHOT_PROMPT)
```

### Results

```python
zero_shot_output = "Billing" 
few_shot_output = "Billing"

print("Zero-shot output:", repr(zero_shot_output))
print("Few-shot output:", repr(few_shot_output))
```

## Part B — Chain-of-Thought (CoT) Prompting (Arithmetic Word Problem)

```python
ZERO_SHOT_MATH_PROMPT = """
A store sells notebooks for $4 each and pens for $2 each.
You buy 3 notebooks and 5 pens, and you pay with a $50 bill.
How much change do you get? Return ONLY the numeric answer.
"""

COT_MATH_PROMPT = """
A store sells notebooks for $4 each and pens for $2 each.
You buy 3 notebooks and 5 pens, and you pay with a $50 bill.

Solve step by step and then give the final numeric answer at the end.
"""

print("ZERO-SHOT MATH PROMPT:\n")
print(ZERO_SHOT_MATH_PROMPT)
print("\nCoT MATH PROMPT:\n")
print(COT_MATH_PROMPT)
```

### Results

```python
zero_shot_math_output = "28"  
cot_math_output = """
3 notebooks × $4 = $12
5 pens × $2 = $10
Total cost = $12 + $10 = $22
Change = $50 − $22 = $28

28"""

print("Zero-shot math output:", repr(zero_shot_math_output))
print("CoT math output:", repr(cot_math_output))
```

## Optional: Running prompts programmatically via API

```python
import os
from openai import OpenAI

client = OpenAI(api_key=os.environ["OPENAI_API_KEY"])

def call_llm(prompt: str, model: str = "gpt-4o-mini") -> str:
    resp = client.chat.completions.create(
        model=model,
        messages=[{"role": "user", "content": prompt}],
        temperature=0,
    )
    return resp.choices[0].message.content.strip()

zero_shot_output = call_llm(ZERO_SHOT_PROMPT)
few_shot_output = call_llm(FEW_SHOT_PROMPT)
zero_shot_math_output = call_llm(ZERO_SHOT_MATH_PROMPT)
cot_math_output = call_llm(COT_MATH_PROMPT)

print(zero_shot_output, few_shot_output, zero_shot_math_output, cot_math_output)
```
